## Let's import Datasets and Analyse them

### Importing covid19 dataset

```python
covid19_confirmed_csv = pd.read_csv("Covid19_Confirmed_dataset.csv")

pd.set_option('display.max_rows', covid19_confirmed_csv.shape[0]+1)
```

### Delete the useless columns

From dataframe information, we can see that Lat and Long are currently not useful for this analysis. So we can delete these columns

```python
covid19_confirmed_csv.drop(['Lat', 'Long'], axis=1, inplace=True)
```

```python
covid19_confirmed_csv.head(10)
```

### Aggregating the rows by the country

```python
covid19_confirmed_agg = covid19_confirmed_csv.groupby("Country/Region").sum()
```

```python
covid19_confirmed_agg.head()
```

```python
covid19_confirmed_agg.shape
```

### Visualizing data related to a country for example China
visualization always helps for better understanding of our data.

```python
covid19_confirmed_agg.loc["China"].plot()
covid19_confirmed_agg.loc["India"].plot()
covid19_confirmed_agg.loc["Spain"].plot()
plt.legend()
```

```python
## Plot partial data i.e. say first 10 days
covid19_confirmed_agg.loc['China'][:10].plot()
```

#### Now we wants to see how much is the infaction rate in an country then we can take the first derivative of the cumulative figures in our dataset

This plot shows infection rate changes day by day

```python
covid19_confirmed_agg.loc["China"].diff().plot()
```

### Calculating a good measure 
we need to find a good measure reperestend as a number, describing the spread of the virus in a country. 

### Lets's find maxmimum infection rate for China

```python
covid19_confirmed_agg.loc["China"].diff().max()
```

We can see that in china, in one day 15136 new infection cases was found.

```python
covid19_confirmed_agg.loc["India"].diff().max()
```

### Let's now find maximum infection rate for all of the countries. 

```python
countries = list(covid19_confirmed_agg.index)
max_infection_rates = []

for country in countries:
    max_infection_rates.append(covid19_confirmed_agg.loc[country].diff().max())
```

```python
covid19_confirmed_agg['max_infection_rate'] = max_infection_rates
```

```python
covid19_confirmed_agg.head()
```

### Now we can create a new dataframe with only needed column

```python
corona_data = pd.DataFrame(covid19_confirmed_agg["max_infection_rate"])
```

```python
corona_data.head()
```

## Let's join both dataframes to make a master dataset

```python
master_data = corona_data.join(happiness_report_csv, how="inner")
master_data.head()
```

### Plotting GDP vs maximum Infection rate

```python
x = master_data["GDP per capita"]
y = master_data["max_infection_rate"]
sns.scatterplot(x,np.log(y))
```

```python
sns.regplot(x,np.log(y))
```

We can see that there is a positive correlation in maximum infection and GDP per capita

### Plotting Social support vs maximum Infection rate

### Plotting Healthy life expectancy vs maximum Infection rate

### Plotting Freedom to make life choices vs maximum Infection rate
